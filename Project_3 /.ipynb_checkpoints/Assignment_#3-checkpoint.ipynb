```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1
# Create a time-series plot showing daily total revenue for three different countries with 
# three countries in one plot

df_daily_countries_revenue_data = pd.read_csv("Task-1-Data.csv")

df_pivot = df_daily_countries_revenue_data.pivot(index='payment_date', columns='country', values='daily_total_revenue')

plt.figure(figsize=(10, 6))
df_pivot.plot()
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.title('Daily Total Revenue for Netherlands, Mexico, and Canada')
plt.legend()

plt.show()
```

```python
# Task 2
# Carry out a sentiment analysis on movie descriptions adding another column to the data frame that provides 
# a sentiment classification. 


df_film_description_data = pd.read_csv("Task-2-Data.csv")

positive_words = [
    'good', 
    'great', 
    'excellent',
    'wonderful', 
    'awesome',
    'amazing',
    'brilliant',
    'captivating',
    'charming',
    'delightful',
    'enjoyable',
    'excellent',
    'fantastic',
    'heartwarming',
    'impressive',
    'inspiring',
    'lovely',
    'masterpiece',
    'outstanding',
    'remarkable',
    'riveting',
    'superb',
    'thrilling',
    'wonderful']

negative_words = [
    'awful', 
    'horrible',
    'bad',
    'boring',
    'disappointing',
    'dreadful',
    'dull',
    'frustrating',
    'lousy',
    'mediocre',
    'poor',
    'terrible',
    'tragic',
    'unimpressive',
    'uninspiring',
    'unpleasant',
    'unwatchable',
    'weak',
    'worse',
    'worst',
    'worthless']

sentiments = []
for description in df_film_description_data['description']:
    positive_count = sum(1 for word in positive_words if word in description.lower())
    negative_count = sum(1 for word in negative_words if word in description.lower())

    sentiment = 'Positive' if positive_count > negative_count else 'Negative' if negative_count > positive_count else 'Neutral'
    sentiments.append(sentiment)

df_film_description_data['Sentiment_Classification'] = sentiments

print(df_film_description_data)
```

```python
# Task 3
# Create a visual in the notebook that shows how much revenue is coming from different countries to the DVD store.

df_countries_revenue_data = pd.read_csv("Task-3-Data.csv")

country_revenue = df_countries_revenue_data.groupby('country')['total_revenue'].sum()
plt.figure(figsize=(20, 6))
country_revenue.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.title('Revenue by Country')
plt.show()
```

```python
# Task 4
# Do an analysis to show which actors are bringing in the most revenue (top 10 ranked)

df_actors_revenue_data = pd.read_csv("Task-4-Data.csv")

plt.figure(figsize=(10, 6))
plt.barh(df_actors_revenue_data['first_name'] + ' ' + df_actors_revenue_data['last_name'], df_actors_revenue_data['total_revenue'])
plt.xlabel('Total Revenue')
plt.ylabel('Actor Name')
plt.title('Top 10 Actors by Revenue')
plt.gca().invert_yaxis() 
plt.show()
```

```python
# Task 5
# Create a new data-frame where you classify some different countries into some categories that make sense to you
# (8 countries min, 20 max) in this df there will be two columns, one column will give the country name, 
# the other column will give the classification category.


countries_data = {
    'country': ['United States', 'United Kingdom', 'Canada', 'Australia', 'Germany', 'France', 'Brazil', 'India', 'China', 'Japan'],
    'Geographical Location': ['North America', 'Europe', 'North America', 'Oceania', 'Europe', 'Europe', 'South America', 'Asia', 'Asia', 'Asia']
}

df = pd.DataFrame(countries_data)

location_counts = df['Geographical Location'].value_counts()

plt.pie(location_counts.values, labels=location_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Countries by Geographical Location')
plt.axis('equal')
plt.show()
```

```python
# Task 6
# Use an appropriate merge to merge that new data-frame from the last step with a dataframe from the dvd db, 
# that gives the countries, so now countries are connected with their classification. 
# (no need to add this table back to the postgresql db, just doing it in the notebook is fine)

dvd_df = pd.read_csv('Task-5-Data.csv')
merged_df = df.merge(dvd_df, on='country', how='left')

print(merged_df)
```
